```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder #import the libraries
```

```python
df = pd.read_csv("play_tennis.csv") #import the dataset
```

```python
df.head()
```

```python
model = DecisionTreeClassifier()
```

```python
new_data = df.drop("day",axis = "columns") #drop the day column
```

```python
le = LabelEncoder()      
```

```python
columns = new_data.columns #find the columns in the dataset 
```

```python
for column in columns:
  new_data[column] = le.fit_transform(df[column])#change the text to integer
```

```python
new_data.head()
```

```python
x = new_data.drop("play",axis = "columns")#create  the x data by droping label
y = new_data["play"]                      #create the label data
x.head()
```

```python
y.head()
```

```python
model.fit(x,y) #train the model
```

Here in the    
1st column :- **Overcast:0,Rain:1,Sunny:2**,   
2nd column :- **Cool:0,Hot:1,Mild:2**,   
3rd column :- **Normal:1,High:0**,   
4th column :- **Weak:1,Strong:0**,        
5th column :-**No:0,Yes:1**

```python
prediction =  {1:"Yes",0:"No"} # craete a for text output for number
```

Prediction for **[sunny,Mild,Normal,Weak]** True output is **Yes**


```python
output1 = model.predict([[2,2,1,1]])[0] #output for [sunny,Mild,Normal,Weak]
prediction[output1]
```

Prediction for **[Rain,Hot,Normal,Weak]** True output is **No**

```python
output2 = model.predict([[1,1,1,1]])[0] #output for [Rain,Hot,Normal,Weak] 
prediction[output2]
```

Prediction for **[Overcast,Mild,Normal,Strong]** True output is **Yes**

```python
output3 = model.predict([[0,2,1,0]])[0] #output for [Overcast,Mild,Normal,Strong] 
prediction[output3]
```

**SO All above three Unseen predictions are Correct**

```python
df_19 = pd.read_csv("covid19.csv") #import the  csv file
df_19.head()
```

```python
new = df_19
```

```python
columns_19 = new.columns
```

```python
le = LabelEncoder()
for column in columns_19:
  df_19[column] = le.fit_transform(new[column]) #change the text columns to integer column
```

```python
df_19.head()
```

```python
x = df_19.drop("COVID-19",axis = "columns") #create the x data by dropping label column(Covid-19)
y = df_19["COVID-19"]                       #create the label column
x.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)#split the train ,test data
```

```python
model_19 = DecisionTreeClassifier()
```

```python
model_19.fit(x_train,y_train)#train the model
```

**ACCURACY OF THE MODEL TO TEST DATA**

```python
model_19.score(x_test,y_test) #accuracy score for test data
```
